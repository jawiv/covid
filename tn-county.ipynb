```python
import pandas as pd

import plotly.express as px
import plotly.graph_objects as go
import cufflinks as cf

with open("../covid-19-data/us-counties.csv") as data:
    county_data = pd.read_csv(data)

nashville_msa_fips = [47015, # Cannon
                      47021, # Cheatham
                      47037, # Davidson
                      47043, # Dickson
                      47081, # Hickman
                      47111, # Macon
                      47119, # Maury
                      47147, # Robertson
                      47149, # Rutherford
                      47159, # Smith
                      47165, # Sumner
                      47169, # Trousdale
                      47187, # Williamson
                      47189, # Wilson
                     ]

county_data['date'] = pd.to_datetime(county_data['date'], format="%Y-%m-%d")
tn_county_data = county_data[(county_data.fips > 47000) & (county_data.fips < 48000)]
msa_data = county_data[county_data.fips.isin(nashville_msa_fips)]
#tn_county_data.set_index("date", inplace=True)
#tn_county_data.head()



pos_tests = tn_county_data.pivot(index='date', columns='county', values='cases')
msa_pos_tests = msa_data.pivot(index='date', columns='county', values='cases')
deaths = tn_county_data.pivot(index='date', columns='county', values='deaths')
msa_deaths = msa_data.pivot(index='date', columns='county', values='deaths')
```

## STATE OF TENNESSEE

```python
fig = pos_tests.iplot(asFigure=True, 
                      kind="bar", 
                      barmode="stack", 
                      xTitle="Date", 
                      yTitle="Cumulative Positive Tests", 
                      title="Cumulative Positive Tests by County - All Tennessee",
                      vspan={'x0':'2020-04-27','x1':'2020-05-05','color':'rgba(30,30,30,0.3)','fill':True,'opacity':.4})
fig.layout.yaxis.tickformat = '0,0'
#fig.update_yaxes(range=[0,1], dtick=0.05)
fig.update_layout(
    autosize=False,
    width=1200,
    height=1000,
    margin=dict(
        l=50,
        r=50,
        b=100,
        t=100,
        pad=4
    ),
)
fig.show()
```

```python
fig = deaths.iplot(asFigure=True, 
                   kind="bar", 
                   barmode="stack", 
                   xTitle="Date", 
                   yTitle="Cumulative Fatalities", 
                   title="Cumulative Fatalities by County - All Tennessee",
                   vspan={'x0':'2020-04-27','x1':'2020-05-05','color':'rgba(30,30,30,0.3)','fill':True,'opacity':.4})
fig.layout.yaxis.tickformat = '0,0'
#fig.update_yaxes(range=[0,1], dtick=0.05)
fig.update_layout(
    autosize=False,
    width=1200,
    height=1000,
    margin=dict(
        l=50,
        r=50,
        b=100,
        t=100,
        pad=4
    ),
)
fig.show()
```
